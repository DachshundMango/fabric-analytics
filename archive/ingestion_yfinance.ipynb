```python
# Import libraries and frameworks

import yfinance as yf
import pandas as pd
```

```python
tickers = ['BHP.AX', 'NG=F', 'AUDUSD=X']

start_date = '2025-01-01'
end_date = '2025-11-30'
```

```python
df_market_full = yf.download(tickers=tickers, start=start_date, end=end_date)

df_market = df_market_full["Close"]

print(df_market.head())
```

```python
df_market.index = pd.to_datetime(df_market.index)

entire_dates = pd.date_range(start='2025-01-01', end='2025-11-30', freq='D')

df_market_final = df_market.reindex(entire_dates)

df_market_final = df_market_final.ffill().bfill()

df_market_final.index.name = 'Date'

df_market_final.rename(columns={'AUDUSD=X': 'AUDUSD_X', 'BHP.AX': 'BHP_AX', 'NG=F': 'NG_F'}, inplace=True)

print(f"final data size: {df_market_final.shape}")
print(df_market_final.head(10))
```

```python
yf_file_path = '../data/yfinance.csv' 
df_market_final.to_csv(path_or_buf=yf_file_path)

print(f"Market status is saved at {yf_file_path}")
```
